```python
import pandas as pd
df = pd.read_csv("passwords.txt", header = None)
df. columns = ["username", "passwords"]
df
```

```python
#Step 2.
total_passwords = len(df)
total_passwords
passwordCount = df["passwords"].value_counts()
passwordCount
passwordMostCount = passwordCount.iloc[0]
passwordMostCount
mostPasswordValue = passwordCount[passwordCount == passwordMostCount]
mostPasswordValue
```

```python
#Step 3
df["password_lower"] = df["passwords"].str.lower()
passwordCount = df["password_lower"].value_counts()
passwordCount
mostCommonIndex = passwordCount.iloc[0]
mostCommonIndex
mostCommonPassword = passwordCount[passwordCount == mostCommonIndex]
mostCommonPassword
```

```python
#Step 4.
df["password_length"] = df["passwords"].str.len()
shortestPassword = df["password_length"].idxmin()
shortestPassword
shortestPasswordValue = df.loc[shortestPassword, "passwords"]
shortestPasswordValue
shortestlength = df.loc[shortestPassword, "password_length"]
shortestlength
print(shortestPasswordValue, shortestlength)
```

```python
#Step 5.
longestPassword = df["password_length"].idxmax()
longestPassword
longestPasswordValue = df.loc[longestPassword, "passwords"]
longestPasswordValue
longestlength = df.loc[longestPassword, "password_length"]
longestlength
print(longestPasswordValue,longestlength)
```

```python
#Step 6
letterString = df["passwords"].str.cat()
type(letterString)
letterSeries = pd.Series(list(letterString))
letterSeries
letterCounts = letterSeries[letterSeries.str.isalpha()].value_counts()
letterCounts
mostOccurringLetter = letterCounts.index[0]
mostOccurringLetter
print(mostOccurringLetter, letterSeries[letterSeries == mostOccurringLetter].count())
```

```python
#Step 7.
letterString = df["passwords"].str.cat()
type(letterString)
letterSeries = pd.Series(list(letterString))
letterSeries
letterCounts = letterSeries[letterSeries.str.isalpha()].value_counts()
letterCounts
leastOccurringLetter = letterCounts.index[-1]
leastOccurringLetter
leastCount=letterCounts.iloc[-1]
leastCount
print(leastOccurringLetter, leastCount)
```

```python
#Step 8.
df["has_uppercase"] = df["passwords"].str.contains(r"[A-Z]",regex=True)
df
UpperCaseDF = df[df["has_uppercase"]]

if len(UpperCaseDF) > 0:
  print(UpperCaseDF["passwords"].head(20).to_string(index=False))
  if len(UpperCaseDF) > 20:
    print(f" ... and {len(UpperCaseDF) - 20} more")
else:
    print("None Found.")
print(f" Total: {len(UpperCaseDF)}")
print()
```

```python
#Step 9.
df["lowerCase_noNum"] = df["passwords"].str.contains(r"^[a-z]+$", regex=True)
lowercaseDF = df[df["lowerCase_noNum"]]
lowercaseDF
if len(lowercaseDF) > 0:
    print(lowercaseDF["passwords"].head(20).to_string(index=False))
    if len(lowercaseDF) > 20:
        print(f" ... and {len(lowercaseDF) - 20} more")
else:
    print(" None Found.")
print(f" Total: {len(lowercaseDF)}")
print()
```
